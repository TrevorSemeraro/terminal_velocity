```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ref_pareto_front = pd.read_csv("../outputs/msre-ref-4/hall_of_fame.csv")
corr_pareto_front = pd.read_csv("../outputs/max-corr-5/hall_of_fame.csv")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

fig.suptitle("Pareto Fronts of Reference and Correction Models", fontsize=16)

ax1.plot(ref_pareto_front['Complexity'], ref_pareto_front['Loss'])
ax1.set_yscale('log')
ax1.set_title("Reference Model")
ax1.set_ylabel("Loss (Mean Squared Relative Error)")
ax1.set_xlabel("Complexity")
ax1.grid()

ax2.plot(corr_pareto_front['Complexity'], corr_pareto_front['Loss'])
ax2.set_yscale('log')
ax2.set_title("Correction Model")
ax2.set_ylabel("Loss (Max Relative Error)")
ax2.set_xlabel("Complexity")
ax2.grid()

plt.tight_layout()
plt.show()
fig.savefig(f'../images/pareto_front.png', bbox_inches='tight', dpi=300)
```
